```python
import pandas as pd

# Load data guru
df_guru = pd.read_csv("files/data_LD_guru.csv", sep=";")
print("Jumlah data guru:", len(df_guru))

# Deteksi kolom Likert (skala 1-5)
likert_cols = [col for col in df_guru.columns if pd.api.types.is_numeric_dtype(df_guru[col])
               and df_guru[col].min() >= 1 and df_guru[col].max() <= 5]
non_likert_cols = [col for col in df_guru.columns if col not in likert_cols]

# Kolom identitas (5 kolom)
kolom_identitas = non_likert_cols[:5]  # pastikan ini memang kolom identitas

# Kolom pernyataan
kolom_likert = likert_cols
kolom_non_likert_pernyataan = non_likert_cols[5:]

print("Kolom Identitas:", len(kolom_identitas))
print("Kolom Pernyataan Likert:", len(kolom_likert))
print("Kolom Pernyataan Non-Likert:", len(kolom_non_likert_pernyataan))
```

```python
df_struktur_kolom = pd.DataFrame({
    "Identitas Responden (5 Kolom)": pd.Series(kolom_identitas),
    "Pernyataan Likert (36 Kolom)": pd.Series(kolom_likert),
    "Pernyataan Non-Likert (4 Kolom)": pd.Series(kolom_non_likert_pernyataan)
})

df_struktur_kolom
```

```python
siaga = pd.read_csv('files/Siaga_updated.csv', sep=';', encoding='latin1')
penggalang = pd.read_csv('files/Penggalang_updated.csv', sep=';', encoding='latin1')
penegak = pd.read_csv('files/Penegak_updated.csv', sep=';', encoding='latin1')

def ambil_kolom(df):
    kolom = (
        ['Nama', 'Kelas', 'Kategori'] +
        [col for col in df.columns if col.startswith('S ')]
    )
    return df[kolom]

siaga_fix = ambil_kolom(siaga)
penggalang_fix = ambil_kolom(penggalang)
penegak_fix = ambil_kolom(penegak)

ambil_kolom(siaga_fix).head(3) 
```
