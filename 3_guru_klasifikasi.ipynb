```python
import pandas as pd
import numpy as np
```

```python
# 1. KLASIFIKASI SKOR TOTAL GURU

# Load data guru
df_guru = pd.read_csv("files/data_LD_guru.csv", sep=';')
print("Jumlah data guru:", len(df_guru))

# Deteksi kolom Likert (skala 1-5)
likert_cols = [col for col in df_guru.columns if pd.api.types.is_numeric_dtype(df_guru[col])
               and df_guru[col].min() >= 1 and df_guru[col].max() <= 5]
non_likert_cols = [col for col in df_guru.columns if col not in likert_cols]

print(f"\nKolom Likert ({len(likert_cols)}): {likert_cols}")
print(f"Kolom Non-Likert ({len(non_likert_cols)}): {non_likert_cols}")
```

```python
# Ambil hanya kolom Likert
df_likert_guru = df_guru[likert_cols]
print("\nData Likert Guru:")
df_likert_guru.head(3)
```

```python
# Bagi dimensi
dim_skill = likert_cols[0:18]
dim_etika = likert_cols[18:25]
dim_keamanan = likert_cols[25:33]
dim_budaya = likert_cols[33:40]

# Hitung rata-rata per dimensi
df_guru["Skill_Digital"] = df_likert_guru[dim_skill].mean(axis=1).round(3)
df_guru["Etika_Digital"] = df_likert_guru[dim_etika].mean(axis=1).round(3)
df_guru["Keamanan_Digital"] = df_likert_guru[dim_keamanan].mean(axis=1).round(3)
df_guru["Budaya_Digital"] = df_likert_guru[dim_budaya].mean(axis=1).round(3)

print("\nSkor Literasi Digital Guru per Dimensi :")
df_guru[["Skill_Digital", "Etika_Digital", "Keamanan_Digital", "Budaya_Digital"]].head()
```

```python
# Hitung total dan mean LD
df_guru["Total_LD"] = df_guru[
    ["Skill_Digital","Etika_Digital","Keamanan_Digital","Budaya_Digital"]
].sum(axis=1).round(3)

df_guru["Mean_LD"] = df_guru[
    ["Skill_Digital","Etika_Digital","Keamanan_Digital","Budaya_Digital"]
].mean(axis=1).round(3)

print("\nSkor Literasi Digital Guru Keseluruhan:")
df_guru[["Total_LD", "Mean_LD"]].head()
```

```python
# 2. KLASIFIKASI SKOR TOTAL

# Cek Trisection
q33 = df_guru["Mean_LD"].quantile(0.33).round(3)
q66 = df_guru["Mean_LD"].quantile(0.66).round(3)

print("\nCek Trisection Mean_LD Guru:")
print("\nNilai Mean_LD pada kuartil 33%  :", q33)
print("Nilai Mean_LD pada kuartil 66%  :", q66) 
```

```python
# Klasifikasi guru
def klasifikasi_ld(x):
    if x <= q33:
        return 'Rendah'
    elif x <= q66:
        return 'Sedang'
    else:
        return 'Tinggi'

df_guru["Level_LD"] = df_guru["Mean_LD"].apply(klasifikasi_ld)

print("\nKlasifikasi Level Literasi Digital Guru:")
df_guru[["NAMA", "Mean_LD", "Level_LD"]].head()
```

```python
# 3. JUMLAH GURU PER KLASIFIKASI LEVEL LD

# Hitung guru
jumlah_guru = df_guru['Level_LD'].value_counts().sort_index()
persentase = (jumlah_guru / jumlah_guru.sum() * 100).round(2)

tabel_ld = pd.DataFrame({
    'Level': jumlah_guru.index,
    'Jumlah Guru': jumlah_guru.values,
    'Persentase (%)': persentase.values
})

print("\nTabel Distribusi Tingkat Literasi Digital Guru")
tabel_ld
```

```python
# CEK OUTPUT DATA GURU
df_guru[[
    "NAMA",
    "Asal Instansi",
    "Total_LD",
    "Mean_LD",
    "Level_LD"
]].head()
```

```python
# Klasifikasi guru
def klasifikasi_ld(x):
    if x <= q33:
        return 'Rendah'
    elif x <= q66:
        return 'Sedang'
    else:
        return 'Tinggi'

df_guru["Level_LD"] = df_guru["Mean_LD"].apply(klasifikasi_ld)

print("\nKlasifikasi Level Literasi Digital Guru:")
df_guru[["Mean_LD", "Level_LD"]].head()
```

```python
# pilih kolom untuk disimpan
df_save_guru = df_guru[[
    "NAMA",
    "Asal Instansi",
    "Total_LD",
    "Mean_LD",
    "Level_LD"
]]

df_save_guru.to_csv(
    "files/data_guru_klasifikasi.csv", 
    index=False, 
    float_format="%.3f"
    )

print("\nData guru berhasil disimpan")
```
